```python
%matplotlib inline

import numpy as np
import pandas as pd
from IPython.display import display
import math

# read training data
data = pd.read_csv("dataset/train.csv")

# print some basic dataset characteristics
print("Training data dimensions: {} x {}".format(*data.shape))
display(data.head())
display(data.describe())

#convert categorical features into numeric


from pandas.plotting import scatter_matrix
# Produce a scatter matrix for each pair of features in the data
#scatter_matrix(data, alpha = 0.3, figsize = (14,8), diagonal = 'kde');

import seaborn as sbrn
import matplotlib.pyplot as plt


#convert categorical columns into numeric
all_cols = data.columns
numeric_cols = data._get_numeric_data().columns
categorical_cols = list(set(all_cols) - set(numeric_cols))
for cat_col in categorical_cols:
    data[cat_col] = pd.Categorical(data[cat_col]).codes

n_of_columns = data.shape[1]
batch_size = 10
n_of_batches = int(math.ceil(n_of_columns / batch_size))
print("Number of batches={}".format(n_of_batches))
for i in range(n_of_batches):
    index_from = i * batch_size
    index_to = (n_of_columns - 1) if (i + 1) * batch_size >= (n_of_columns - 1) else (i + 1) * batch_size
    print("Batch number:{}, index_from={}, index_to={}".format(i, index_from, index_to))
    if index_to - index_from <= 1:
        break
        print("breaking")
    indices = list(range(index_from, index_to))
    # adding the last column = price
    indices.append(n_of_columns - 1)
    
    data_batch = data.iloc[:, indices]
    print("data_batch.keys()={}".format(data_batch.keys()))
    
    #display(data_batch.head())
    #display(data_batch.describe())
  
    fig, ax = plt.subplots(figsize=(15,10))
    sbrn.heatmap(data_batch.corr(), annot=True, ax=ax)
    #sbrn.heatmap(data_batch.corr(), annot=True)
    
    scatter_matrix(data_batch, alpha = 0.3, figsize = (15,13), diagonal = 'kde');
    
    plt.show()
```
